```python
data_folder='/Users/albitraverso/WorkspaceMaastro/P0484_GTV_NSCLC/analysis/'
roi_name='GTV'
```

```python
patient_list=list()
csv_list=list()
df_list=list()
```

```python
excluded_list=[]
```

```python
if os.name == 'nt':
    import win32api, win32con
def file_is_hidden(p):
    if os.name== 'nt':
        attribute = win32api.GetFileAttributes(p)
        return attribute & (win32con.FILE_ATTRIBUTE_HIDDEN | win32con.FILE_ATTRIBUTE_SYSTEM)
    else:
        return p.startswith('.') #linux-osx
    import os
if os.name == 'nt':
    import win32api, win32con
def file_is_hidden(p):
    if os.name== 'nt':
        attribute = win32api.GetFileAttributes(p)
        return attribute & (win32con.FILE_ATTRIBUTE_HIDDEN | win32con.FILE_ATTRIBUTE_SYSTEM)
    else:
        return p.startswith('.') #linux-osx
```

```python
for p in sorted(os.listdir(data_folder)):
    if not file_is_hidden(p):
        patient_list.append(p)
    if p in excluded_list:
      print("Excluding %s" %p)
      patient_list.remove(p)
print(len(patient_list))
print(patient_list)
```

```python
for p in patient_list:
    csv_list.append(os.path.join(data_folder,p,roi_name,"%s_%s_radiomic_features.csv" % (p,roi_name)))
print(len(csv_list))
print(csv_list)
```

```python
for csv in csv_list:
    df = pd.read_csv(csv, index_col=None, header=0)
    df_list.append(df)
print(len(df_list))
frame = pd.concat(df_list, axis=0, ignore_index=True)
frame.insert(0,column="ID",value=patient_list)
```

```python
#Remove all the columns that are not needed (diagnostics)
frame_polished = frame[frame.columns.drop(list(frame.filter(regex='diagnostics_')))]
```
